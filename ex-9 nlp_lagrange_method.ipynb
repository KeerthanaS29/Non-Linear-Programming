```python
import sympy as sp

# Step 1: Take user inputs
num_vars = int(input("Enter the number of variables: "))
num_constraints = int(input("Enter the number of constraints: "))

# Step 2: Create variable symbols
var_names = input(f"Enter {num_vars} variable names separated by space (example: x y z): ").split()
vars = sp.symbols(var_names)

# Step 3: Create lambda (Lagrange multiplier) symbols
lambdas = sp.symbols(f'lam0:{num_constraints}')

# Step 4: Input the objective function
f_input = input("Enter the objective function f(...): ")
f = sp.sympify(f_input)

# Step 5: Input the constraint functions
constraints = []
for i in range(num_constraints):
    g_input = input(f"Enter constraint {i+1} function g{i+1}(...)=0: ")
    constraints.append(sp.sympify(g_input))

# Step 6: Build the Lagrangian
L = f
for i in range(num_constraints):
    L += lambdas[i] * constraints[i]

print("\nLagrangian L(...):", L)

# Step 7: Compute partial derivatives
equations = []
for v in vars:
    equations.append(sp.diff(L, v))  # âˆ‚L/âˆ‚(variable) = 0
for lam in lambdas:
    equations.append(sp.diff(L, lam))  # âˆ‚L/âˆ‚(lambda) = 0

# Step 8: Solve the system
all_symbols = list(vars) + list(lambdas)
solutions = sp.solve(equations, all_symbols, dict=True)

# Step 9: Output
print("\nSolutions (critical points):")
for idx, sol in enumerate(solutions):
    print(f"\n--- Solution {idx+1} ---")
    for sym in all_symbols:
        print(f"{sym} = {sol[sym]}")
    
    # Step 10: Find function value at this critical point
    f_value = f.subs(sol)
    print(f"\nFunction value at this point: f = {f_value}")
    
    # Step 11: Check nature of critical point
    # Build Hessian matrix of original function f
    hessian = sp.hessian(f, vars)
    
    # Substitute the solution into Hessian
    hessian_at_sol = hessian.subs(sol)
    
    # Find eigenvalues
    eigenvals = hessian_at_sol.eigenvals()
    eigenvalues = list(eigenvals.keys())
    
    print("\nEigenvalues of Hessian at this point:", eigenvalues)
    
    # Analyze eigenvalues
    if all(ev > 0 for ev in eigenvalues):
        print("Conclusion: Local Minimum âœ…")
    elif all(ev < 0 for ev in eigenvalues):
        print("Conclusion: Local Maximum âœ…")
    else:
        print("Conclusion: Saddle Point âŒ")

        
```

```python
import sympy as sp

# Step 1: Take user inputs
num_vars = int(input("Enter the number of variables: "))
num_constraints = int(input("Enter the number of constraints: "))

# Step 2: Define variable symbols
var_names = input(f"Enter {num_vars} variable names separated by space (e.g., x1 x2 x3): ").split()
vars = sp.symbols(var_names)

# Step 3: Define Lagrange multipliers
lambdas = sp.symbols(f'lam0:{num_constraints}')

# Step 4: Input the objective function
print("\nðŸ‘‰ Use '*' for multiplication (e.g., 3*x1, not 3x1)")
f_input = input("Enter the objective function f(...): ")
f = sp.sympify(f_input)

# Step 5: Input the constraint equations
constraints = []
for i in range(num_constraints):
    g_input = input(f"Enter constraint {i+1} in the form g{i+1}(...) = 0: ")
    constraints.append(sp.sympify(g_input))

# Step 6: Build the Lagrangian function (CORRECTED with minus sign)
L = f
for i in range(num_constraints):
    L -= lambdas[i] * constraints[i]  # Changed += to -=

print("\nðŸ”§ Lagrangian L(...):", L)

# Step 7: Compute the system of equations from the gradient
equations = [sp.diff(L, v) for v in vars] + [sp.diff(L, lam) for lam in lambdas]

# Step 8: Solve the system
all_symbols = list(vars) + list(lambdas)
solutions = sp.solve(equations, all_symbols, dict=True)

# Step 9: Output results
if not solutions:
    print("\nâŒ No solutions found. Please check your input.")
else:
    print("\nâœ… Solutions (critical points):")
    for idx, sol in enumerate(solutions):
        print(f"\n--- Solution {idx+1} ---")
        for sym in all_symbols:
            print(f"{sym} = {sp.simplify(sol[sym])}")
        
        # Step 10: Evaluate the function at this solution
        f_value = f.subs(sol)
        print(f"\nðŸŽ¯ Objective function value at this point: f = {f_value}")

        # Step 11: Interpret the Lagrange multipliers
        for i in range(num_constraints):
            lam_val = sol[lambdas[i]]
            if lam_val > 0:
                interpretation = "active and increasing f"
            elif lam_val < 0:
                interpretation = "active and decreasing f"
            else:
                interpretation = "inactive (no effect on f)"
            print(f"Î»{i} = {sp.simplify(lam_val)} â†’ Constraint {i+1} is {interpretation}")

        # Step 12: Analyze the nature of the critical point
        hessian = sp.hessian(f, vars)
        hessian_at_sol = hessian.subs(sol)

        try:
            eigenvals = hessian_at_sol.eigenvals()
            eigenvalues = list(eigenvals.keys())
            print("\nðŸ“ Eigenvalues of Hessian at this point:", eigenvalues)

            if all(ev > 0 for ev in eigenvalues):
                print("Conclusion: Local Minimum âœ…")
            elif all(ev < 0 for ev in eigenvalues):
                print("Conclusion: Local Maximum âœ…")
            else:
                print("Conclusion: Saddle Point âŒ")
        except:
            print("âš ï¸ Could not determine nature of critical point (symbolic Hessian issue).")
```

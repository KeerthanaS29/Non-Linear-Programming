```python
import math
 
def f(x):
    return -x**2 - 1  # Function to maximize (downward parabola)
 
def golden_section_search_maximize(xL, xU, max_uncertainty=0.25, max_iter=100):
    R = (math.sqrt(5) - 1) / 2  # Golden ratio (~0.618)
    for i in range(max_iter):
        d = R * (xU - xL)
        x1 = xL + d
        x2 = xU - d
        f1 = f(x1)
        f2 = f(x2)
        if f1 > f2:  # Keep the interval where the maximum is
            xL = x2
            x_opt = x1
        else:
            xU = x1
            x_opt = x2
        current_uncertainty = xU - xL
        print(f"Iteration {i + 1}:")
        print(f"  xL = {xL:.5f}, xU = {xU:.5f}")
        print(f"  x1 = {x1:.5f}, f(x1) = {f1:.5f}")
        print(f"  x2 = {x2:.5f}, f(x2) = {f2:.5f}")
        print(f"  d = {d:.5f}, Uncertainty = {current_uncertainty:.5f}")
        print("-" * 50)
        if current_uncertainty <= max_uncertainty:
            break
    return [x_opt, current_uncertainty]
 
# Initial interval
xL = -1.0
xU = 0.75
max_uncertainty = 0.25  # Stopping condition: interval â‰¤ 1/4
max_iter = 100
 
result = golden_section_search_maximize(xL, xU, max_uncertainty, max_iter)
print("\nFinal Result:")
print(f"Optimal x = {result[0]:.5f}")
print(f"f(x) = {f(result[0]):.5f}")
print(f"Final Uncertainty Interval = {result[1]:.5f}")
```

```python
import math
 
def f(x):
    return x**4 - 14*x**3 + 60*x**2 - 70*x  # Function to minimize
 
def golden_section_search_minimize(xL, xU, max_uncertainty=0.15, max_iter=100):
    R = (math.sqrt(5) - 1) / 2  # Golden ratio (~0.618)
    for i in range(max_iter):
        d = R * (xU - xL)
        x1 = xL + d
        x2 = xU - d
        f1 = f(x1)
        f2 = f(x2)
 
        if f1 < f2:  # Keep the subinterval with the lower value
            xU = x1
            x_opt = x2
        else:
            xL = x2
            x_opt = x1
 
        current_uncertainty = xU - xL
        print(f"Iteration {i + 1}:")
        print(f"  xL = {xL:.5f}, xU = {xU:.5f}")
        print(f"  x1 = {x1:.5f}, f(x1) = {f1:.5f}")
        print(f"  x2 = {x2:.5f}, f(x2) = {f2:.5f}")
        print(f"  Uncertainty = {current_uncertainty:.5f}")
        print("-" * 50)
 
        if current_uncertainty <= max_uncertainty:
            break
 
    return (xL + xU) / 2  # Return midpoint of final interval
 
# Parameters
xL = 0.0
xU = 2.0
max_uncertainty = 0.15
max_iter = 100
 
# Run the search
result = golden_section_search_minimize(xL, xU, max_uncertainty, max_iter)
print("\nFinal Result:")
print(f"Optimal x = {result:.5f}")
print(f"Minimum f(x) = {f(result):.5f}")
```

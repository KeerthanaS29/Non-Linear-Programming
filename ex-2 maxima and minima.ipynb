1. f(x)=10-x^2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
from scipy.misc import derivative

def analyze_function():
    func_str = input("Enter a function of x (e.g., '10 - x**2' or 'x**3 - 2*x'): ")
    
    try:
        f = lambda x: eval(func_str)
        test_val = f(0) 
    except:
        print("Invalid function! Please use Python syntax with variable 'x'.")
        return
    
    try:
        second_deriv = derivative(f, 0.0, n=2, dx=1e-3)
        if second_deriv > 0:
            concavity = "convex (concave up)"
        elif second_deriv < 0:
            concavity = "concave (concave down)"
        else:
            concavity = "undetermined concavity"
    except:
        concavity = "could not determine concavity"
    
    print(f"\nFunction analysis for f(x) = {func_str}")
    print(f"Concavity at x=0: {concavity}")
    
    try:
        f_prime = lambda x: derivative(f, x, dx=1e-5)
        critical_points = []
        
        search_intervals = [(-5, -0.1), (-0.1, 0.1), (0.1, 5)]
        for a, b in search_intervals:
            try:
                res = minimize_scalar(lambda x: abs(f_prime(x)), bounds=(a, b), method='bounded')
                if abs(f_prime(res.x)) < 1e-3:  # If derivative is close to zero
                    critical_points.append(res.x)
            except:
                continue
        critical_points = list(set([round(x, 3) for x in critical_points]))
        
        print("\nCritical points found (where f'(x)≈0):")
        for x in critical_points:
            print(f"x = {x:.3f}, f(x) = {f(x):.3f}")
            try:
                second_deriv = derivative(f, x, n=2, dx=1e-3)
                if second_deriv > 1e-3:
                    print(f"  → Local minimum (f''(x) = {second_deriv:.3f} > 0)")
                elif second_deriv < -1e-3:
                    print(f"  → Local maximum (f''(x) = {second_deriv:.3f} < 0)")
                else:
                    print("  → Test failed (could be saddle point)")
            except:
                print("  → Could not classify (derivative calculation failed)")
        
    except Exception as e:
        print(f"\nError finding critical points: {str(e)}")
    
    plt.figure(figsize=(10, 6))
    
    x_plot = np.linspace(-3, 3, 400)
    y_plot = [f(x) for x in x_plot]
    
    plt.plot(x_plot, y_plot, label=f"f(x) = {func_str}")
    
    for x in critical_points:
        plt.scatter(x, f(x), color='red', s=100, 
                  label=f'Critical point at x={x:.2f}')
    
    plt.grid(True)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(f"Function Analysis: {func_str}\nConcavity: {concavity}")
    plt.legend()
    plt.show()

analyze_function()
```

3. f(x)= 2x^3-3x^2

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

def is_convex(f, x_range=(-5, 5), num_points=100):
    """Check if f(x) is convex (PSD Hessian) in the given range."""
    x_test = np.linspace(x_range[0], x_range[1], num_points)
    for x in x_test:
        f_xx = derivative(f, x, n=2, dx=1e-5)
        if f_xx < -1e-6: 
            return False, x
    return True, None

def analyze_and_plot():
    func_str = input("Enter f(x) (e.g., '2*x**3 - 3*x**2'): ")
    try:
        f = lambda x: eval(func_str)
        f(0) 
    except:
        print("Invalid function! Use Python syntax with 'x'.")
        return

    convex, violator = is_convex(f)
    print(f"\nAnalysis of f(x) = {func_str}")
    if convex:
        print("✅ The function is convex (PSD) everywhere in the tested range.")
    else:
        print(f"❌ The function is NOT convex (f''(x) < 0 near x ≈ {violator:.2f}).")

    x_vals = np.linspace(-0.5, 1.5, 400)
    convex_mask = [derivative(f, x, n=2, dx=1e-5) >= -1e-6 for x in x_vals]

    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, [f(x) for x in x_vals], label=f'$f(x) = {func_str}$', color='blue')
    
    plt.fill_between(x_vals, [f(x) for x in x_vals], 
                    where=convex_mask, color='green', alpha=0.3,
                    label='Convex region (f\'\'(x) ≥ 0)')

    f_prime = lambda x: derivative(f, x, dx=1e-5)
    critical_points = []
    for x in x_vals[::50]:  # Coarse search
        try:
            root = minimize_scalar(lambda x: abs(f_prime(x)), bracket=(x-0.5, x+0.5)).x
            if abs(f_prime(root)) < 1e-5:
                critical_points.append(root)
        except:
            pass
    critical_points = list(set([round(x, 3) for x in critical_points]))

    for x_crit in critical_points:
        plt.scatter(x_crit, f(x_crit), color='red', s=100,
                   label=f'Critical point at x={x_crit:.2f}')

    plt.title(f"Convexity Analysis: $f(x) = {func_str}$\n"
              f"Convex (PSD) everywhere? {'Yes' if convex else 'No'}")
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

    if not convex:
        print("\nTo make the function convex:")
        print("1. Restrict domain to where f''(x) ≥ 0.")
        print("2. Add a regularization term (e.g., use f(x) + λx² where λ ≥ |min(f''(x))|).")

analyze_and_plot()
```
